# **Question - 1**


**Q.1 (a) Trapezoidal Rule**

```python
import math
import numpy as np

def integrand(x):
  return math.pow(x,4) - 2*x + 1

def trapez_integ(integ_func, low_lim, high_lim, n_slices):
  step = (high_lim - low_lim)/n_slices
  sum = (integ_func(low_lim) + integ_func(high_lim))/2.
  for i in range(1,n_slices):
    sum += integ_func(low_lim + step*i)
  return sum*step

result_trap = trapez_integ(integrand,0,2,10)
error_trap = abs(result_trap - 4.4)/4.4
print(f"Result for given function with 10 slices is {result_trap} with fractional error {error_trap}")
```

**Q.1 (b) Simpson's Rule**

Run part (a) once if not done till now. Since this block uses functions defined in those cells.

```python
# run part (a) above first since it loads modules and defines integrand function

def simpsons_integ(integ_func, low_lim, high_lim, n_slices):
  step = (high_lim - low_lim)/n_slices
  sum = (integ_func(low_lim) + integ_func(high_lim))/3.
  for i in range(1,n_slices):
    if (i%2 == 0):
      sum += (2./3.)*integ_func(low_lim + step*i)
    else:
      sum += (4./3.)*integ_func(low_lim + step*i)
  return sum*step

result_simp = simpsons_integ(integrand,0,2,10)
error_simp = abs(result_simp - 4.4)/4.4
print(f"Result for given function with 10 slices is {result_simp} with fractional error {error_simp}")
```

**Q.2 (a) Adaptive trapezoid rule**

```python
# redefining trapezoid method function from Q.1 (a) so that this block works independently

import math
import time
import numpy as np

def integrand_2(x):
  return math.pow(math.sin(math.sqrt(100*x)),2)

def trapez_integ(integ_func, low_lim, high_lim, n_slices):
  step = (high_lim - low_lim)/n_slices
  sum = (integ_func(low_lim) + integ_func(high_lim))/2.
  for i in range(1,n_slices):
    sum += integ_func(low_lim + step*i)
  return sum*step

accuracy = 1e-6
slices = 1
start = time.time()

I_1 = trapez_integ(integrand_2,0,1,slices)

print(f"Given integral for 1 slice is {I_1}")

while (True):
  slices *= 2
  I_2 = trapez_integ(integrand_2,0,1,slices)
  err_latest = abs((I_2-I_1)/3.)
  print(f"Integral for {slices} slices is {I_2} with error {err_latest}")
  if (err_latest<accuracy):
    print()
    print(f'time taken = {time.time()-start} s')
    break
  I_1 = I_2
```

**Q.2 (b) Romberg integration**

Run part (a) first since it defines the trapezoidal function

```python
# run part (a) first, since it defines the trapezoidal integration function

accuracy = 1e-6
Rom = [[],[]]   # elements of Rom are the individual rows of our triangle
slices = 1
count = 0

start = time.time()
# Not that for our code, we start the 2D list from 0,0 for the first step due to conventional array indexing
Rom[0].append(trapez_integ(integrand_2,0,1,slices))
slices *= 2
count += 1
Rom[1].append(trapez_integ(integrand_2,0,1,slices))
err = abs((Rom[1][0]-Rom[0][0])/3.)

while (True):
  if (err<accuracy):
    break
  for m in range(0,count):
    R_i_mplus1 = Rom[count][m] + (Rom[count][m]-Rom[count-1][m])/(math.pow(4,m+1)-1)    # power has M=1 since m is actually an index starting from 0
    Rom[count].append(R_i_mplus1)
    err = abs(Rom[count][m]-Rom[count-1][m])/(math.pow(4,m+1)-1)
    if (err<accuracy):
      print(f'accuracy achieved for penultimate result of last row is {err}')
      print(f"most accurate estimate = {R_i_mplus1}")
      print(f'time taken = {time.time()-start}')
      print()
      break
  slices *= 2
  count += 1
  Rom.append([trapez_integ(integrand_2,0,1,slices)])

for i in Rom:
  for j in i:
    print(j,end='  ')
  print()
```

We can clearly see form the time taken by both adaptive trapezoidal rule and romberg method, that the later is faster.

**Q.3 (a) Expression for time period**

We have $E=\frac{1}{2}m\left(\frac{dx}{dt}\right)^2 + V(x)$

$\therefore \frac{dx}{dt} = \pm \sqrt{\frac{2(E-V(x))}{m}}$

where the sign will be -ve for our integration (velocity is in -ve direction). Integrating above from t=0 to $T/4$ and x=a to x=0, where $T$ is the time period, we have

$\frac{T}{4} = -\int_a^0 \sqrt{\frac{m}{2(E-V(x))}}\, dx \rightarrow T = 4\int_0^a \sqrt{\frac{m}{2(V(a)-V(x))}}\, dx$

```python
import numpy as np
import matplotlib.pyplot as plt
import math

global mass          # mass of particle
mass = 1

def potential(x,n):
  return np.pow(x,n)

def my_integrand(x,a,n):
  V_0 = potential(a,n)      # a is the amplitude
  V_x = potential(x,n)
  return np.sqrt(mass/(2.*(V_0-V_x)))

amplitudes = np.arange(0,2,0.05)

x, w = np.polynomial.legendre.leggauss(20) # points and weights for [-1,1]

plt.figure(figsize=(6,6))
periods = []

for amp in amplitudes:
  # mapping gaussian quad points to our domain and adjusting weights
  xp = 0.5*(amp)*x + 0.5*(amp)
  wp = 0.5*(amp)*w
  res = np.sum(wp*my_integrand(xp,amp,4)) # our potential has x^4. we can change n to see behavior for x^2, root(x)
  periods.append(res)

plt.plot(amplitudes,np.array(periods))
plt.xlabel("amplitude a")
plt.ylabel("time period T")
plt.grid(axis="both")

```

**Q.3 (c)** To check the variation of the integral with amplitude a, we can substitute $x=ay$ which gives

$ T = 4\int_0^a \sqrt{\frac{m}{2(V(a)-V(x))}}\, dx = 4\int_0^1 \sqrt{\frac{m}{2(a^4-a^4y^4)}}\, ady$

$ \therefore T = \frac{4}{a}\int_0^1 \sqrt{\frac{m}{2(1-y^4)}}\, dy  T \propto 1/a$

Hence the time period diverges near 0 and decreases with increasing amplitudes.

**Q.4 (a)**

For a general infinitesimal surface element of the sheet, the gravitational force on the point mass is given by

$d\vec{F} = \frac{-Gdm}{(x^2+y^2+z^2)}\hat{F}$

where $\hat{F}$ is a unit vector along the line joining this surface element and the point mass. The component of this force along $\hat{z}$ is

$dF_z=-\frac{G\sigma\ dx\ dy}{(x^2+y^2+z^2)}\cdot \frac{z}{(x^2+y^2+z^2)^{1/2}} = -\frac{G\sigma z\ dx\ dy}{(x^2+y^2+z^2)^{3/2}}$

Integrating throughout the sheet we get

$F_z = -G\sigma z\int \int_{-L/2}^{L/2}\frac{dx\ dy}{(x^2+y^2+z^2)^{3/2}}$

```python
import numpy as np
import matplotlib.pyplot as plt
import math

global G,L,M         # sheet properties and G
L = 10
M = 1e4
G = 6.67e-11
sig = M/(L**2)

def my_integrand1(x,y,z):
  denom = math.pow((x**2 + y**2 + z**2),1.5)
  return 1/denom

heights = np.arange(0,10,0.05)

x, w = np.polynomial.legendre.leggauss(100) # points and weights for [-1,1]
y, w = np.polynomial.legendre.leggauss(100)

plt.figure(figsize=(6,6))
forces = np.zeros(len(heights))

count=0
for z in heights:
  # mapping gaussian quad points to our domain [-L/2,L/2] and adjusting weights
  xp = 0.5*(L)*x
  wpx = 0.5*(L)*w
  yp = 0.5*(L)*y
  wpy = 0.5*(L)*w
  sum = 0
  for i in range(0,len(xp)):
    for j in range(0,len(yp)):
       sum += wpx[i]*wpy[j]*z*my_integrand1(xp[i],yp[j],z)
  forces[count]=sum
  count += 1

plt.plot(heights,G*sig*forces)
plt.xlabel("height z")
plt.ylabel("Force in negative z direction")
plt.grid(axis="both")
```

We see that for very small z, the integrand becomes

$F_z \approx \frac{z}{(x^2+y^2)^{3/2}}$

which is very sharply peak near (x,y)=(0,0). Now, Gaussian quadrature uses fixed node points to evaluate the integral. Therefore for very small z, the high contribution coming from the central peak may not be accounted for, therefore giving zero as the value of the integral.

To solve this issue, one can increase the number of nodes near the origin. Or use some method that increases the density of sample points near (x,y)=(0,0). A change of integration variables to remove this singularity will also work.

**Q.5 (c) Plotting diffraction pattern**

Since there are 10 slits in total, we can calculate he total width of grating

$w_{total} = 20\times 10^{-6} \cdot 10 = 2\times 10^{-4} $ m

therefore our integral becomes

$I(x) = |\int_{-w/2}^{w/2}\sqrt{q(u)}e^{i2\pi xu/\lambda f}\, du|^2$

where $w/2 = 10^{-4}$ m

```python
# run part (a) once before this

import cmath
import numpy as np
import matplotlib.pyplot as plt

global Lambda_,Focal_L,width    # defininf wavelength of light, focal length of lens and width of grating
Lambda_ = 5e-7
Focal_L = 1
width = 2e-4

def integrand_diff(x,u):
  temp = cmath.exp(2j*cmath.pi*x*u/(Lambda_*Focal_L))
  return cmath.sqrt(trans_func(u))*temp

# we use adaptive simpson's rule from Q.2 part (c)
# we define it here again for independence of this block

def simpsons_integ(integ_func, x, low_lim, high_lim, n_slices):
  step = (high_lim - low_lim)/n_slices
  sum = (integ_func(x,low_lim) + integ_func(x,high_lim))/3.
  for i in range(1,n_slices):
    if (i%2 == 0):
      sum += (2./3.)*integ_func(x,low_lim + step*i)
    else:
      sum += (4./3.)*integ_func(x,low_lim + step*i)
  return sum*step

# screen is 0.1 m wide
screen_points = np.arange(-0.05,0.05,0.0001)  # in m
Intensity_1 = np.zeros(len(screen_points))
accuracy = 1e-8

counter = 0
for x in screen_points:
  slices = 2
  I_1 = simpsons_integ(integrand_diff,x,-width/2,width/2,slices)
  while (True):
    slices *= 2
    I_2 = simpsons_integ(integrand_diff,x,-width/2,width/2,slices)
    err_latest = abs((I_2-I_1)/15.)
    if (err_latest<accuracy):
      break
    I_1 = I_2
  Intensity_1[counter] = abs(I_2)*abs(I_2)
  counter += 1

plt.figure(figsize=(4,4))
plt.plot(screen_points, Intensity_1)
plt.grid(axis='both')
plt.xlabel("x")
plt.ylabel("Intensity")
plt.show()
```

```python
import cmath
import math

def z_k(k,N):
  return cmath.exp(2j*cmath.pi*k/N)

def func(zk):
  return cmath.exp(2*zk)

def calc_mth_deriv(function,m,N):
  sum = 0
  for k in range(0,N):
    sum += function(z_k(k,N))*cmath.exp(-2j*cmath.pi*k*m/N)
  return math.factorial(m)*sum/N

for m in range(1,21):
  res = calc_mth_deriv(func,m,10000)
  print(f"{m} th derivative of the given function at z=0 is {res}. Actual value is {int(math.pow(2,m))}. Relative error = {abs((res/math.pow(2,m))-1)}")
```

# **Question - 7**

**Q.7 (a)** We use forward difference method for leading edge of map (where x = 0 or y = 0), backward difference for trailing edge (where x and y are max) and central difference method for other points in between.

```python
import requests
import numpy as np
from io import StringIO

url = "https://raw.githubusercontent.com/ArnavK04/PH354-Assignments/refs/heads/main/altitude.txt"

response = requests.get(url)
response.raise_for_status()

alt_data = np.loadtxt(StringIO(response.text))

def forward_diff_y(array,i,j,step):
  return (array[i,j+1]-array[i,j])/step

def forward_diff_x(array,i,j,step):
  return (array[i+1,j]-array[i,j])/step

def backward_diff_y(array,i,j,step):
  return (array[i,j]-array[i,j-1])/step

def backward_diff_x(array,i,j,step):
  return (array[i,j]-array[i-1,j])/step

def central_diff_x(array,i,j,step):
  return (array[i+1,j]-array[i-1,j])/(2*step)

def central_diff_y(array,i,j,step):
  return (array[i,j+1]-array[i,j-1])/(2*step)

def calc_derivatives(altitudes,h):
  deriv_x = np.zeros(altitudes.shape)
  deriv_y = np.zeros(altitudes.shape)
  Nx,Ny = altitudes.shape
  for i in range(0,Nx):
    for j in range(0,Ny):
      if (i==0):
        deriv_x[i,j] = forward_diff_x(altitudes,i,j,h)
      elif (i==(Nx-1)):
        deriv_x[i,j] = backward_diff_x(altitudes,i,j,h)
      else:
        deriv_x[i,j] = central_diff_x(altitudes,i,j,h)
      if (j==0):
        deriv_y[i,j] = forward_diff_y(altitudes,i,j,h)
      elif (j==(Ny-1)):
        deriv_y[i,j] = backward_diff_y(altitudes,i,j,h)
      else:
        deriv_y[i,j] = central_diff_y(altitudes,i,j,h)
  return deriv_x,deriv_y
```

**Q.7 (b)** Run part (a) once before this. Derivative arrays are calculated there

```python
import math
import numpy as np
import matplotlib.pyplot as plt

phi = math.pi/4
h = 30000

deriv_x,deriv_y = calc_derivatives(alt_data,h)

denominator = np.sqrt(deriv_x**2 + deriv_y**2 + np.ones_like(deriv_x))
numerator = math.cos(phi)*deriv_x + math.sin(phi)*deriv_y

I = numerator/denominator

plt.figure(figsize=(12,6))

plt.imshow(I,cmap='gray')
plt.colorbar()
plt.grid(axis='both')
plt.show()

```
